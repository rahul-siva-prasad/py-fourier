```python
from pyFourier import *
```

```python
sig1 = f_buildSinWave(freq=1)
sig2 = f_buildSinWave(amp = 0.12, freq=20)
sig3 = f_buildSinWave(amp = 0.12, freq=10, phase=90)
dcShift = f_buildDcShift(0.125)
sig4 = sig1+sig2+sig3+dcShift
```

```python
sigs = [sig1,sig2,sig3,dcShift]
```

```python
f_plotGraph(*sigs,title = "Harmonics !!!")
```

```python
f_plotGraph(sig4,title="Resultant wave")
```
